```python
import numpy as np
import scipy
import sklearn
import librosa
import os


dirname = 'D:/Documents/Final Year Project/Wav files/Survey 7/'
```

```python
mylist = os.listdir(dirname)    
mylist.sort(key=lambda f: int(''.join(filter(str.isdigit, f))))
```

```python
mylist = mylist[0:300]
mylist
```

```python
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
```

## Calc Spectral Features for all songs in list

```python
%%time
spectral_list = []
for i in mylist:
    x, sr = librosa.load(dirname+i, offset = 20, duration = 30)
    
    spectral_contrast = librosa.feature.spectral_contrast(x, sr=sr)
    norm_sc = normalize(spectral_contrast, axis=1)
    
    spectral_centroids = librosa.feature.spectral_centroid(x, sr=sr)[0]
    norm_scentroid = normalize(spectral_centroids)
    
    flatness = librosa.feature.spectral_flatness(y=x)
    norm_sf = normalize(flatness.T)
    mean_sc = np.mean(norm_sc, axis=1)
    std_sc = np.std(norm_sc, axis=1)
    spectral_features = [*mean_sc, *std_sc, np.mean(norm_scentroid), np.std(norm_scentroid), np.mean(norm_sf), np.std(norm_sf)]
    spectral_list.append(spectral_features)     #Add to list
```

```python
import csv
data_head = ['Mean_SC0','Mean_SC1','Mean_SC2','Mean_SC3','Mean_SC4','Mean_SC5','Mean_SC6','SD_SC0','SD_SC1','SD_SC2','SD_SC3','SD_SC4','SD_SC5','SD_SC6','MeanCentroid', 'SDCentroid', 'MeanFlatness', 'SDFlatness']
with open("output_spectral.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerow(data_head)
    writer.writerows(spectral_list)
```
