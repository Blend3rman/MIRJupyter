```python
import numpy as np
import scipy
import sklearn
import librosa
import os
from scipy import signal

dirname = 'D:/Documents/Final Year Project/Wav files/Survey 7/'
```

```python
mylist = os.listdir(dirname)    
mylist.sort(key=lambda f: int(''.join(filter(str.isdigit, f))))
```

```python
mylist = mylist[0:300]
mylist
```

```python
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
```

## Calc Spectral Features for all songs in list

```python
%%time
rhythm_list = []
for i in mylist:
    x, sr = librosa.load(dirname+i, offset = 20, duration = 30)
    
    onset_env = librosa.onset.onset_strength(x, sr=sr, hop_length=200, n_fft=2048)
    peaks, _= signal.find_peaks(onset_env, height=1)
    t_peaks = librosa.frames_to_time(peaks, sr=sr, hop_length=200)
    t_diff = [t_peaks[i+1]-t_peaks[i] for i in range(len(t_peaks)-1)]
    
    peak_density = 1/np.mean(t_diff)
    
    tempo = librosa.beat.tempo(x, sr=sr)
    
    rhythm_features = [peak_density, tempo]
    rhythm_list.append(rhythm_features)     #Add to list
```

```python
for i in rhythm_list:
    i[1] = float(i[1])


rhythm_list
```

```python
import csv
data_head = ['PeakDensity', 'Tempo']
with open("output_rhythm.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerow(data_head)
    writer.writerows(rhythm_list)
```
